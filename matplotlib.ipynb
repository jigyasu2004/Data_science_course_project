```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#object oriented method
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])


axes1.plot(x,y)
axes2.plot(y,x)
```

```python
fig,axes = plt.subplots(nrows = 1, ncols = 2)
#plt.tight_layout()
for current_ax in axes:
    current_ax.plot(x,y)
```

```python
axes
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 1,figsize = (10,8))
axes[0].plot(x,y)
axes[1].plot(y,x)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, label = 'x,y')
ax.plot(y,x, label = 'y,x')
ax.legend(loc = 0)
```

```python
fig = plt.figure(figsize = (10,5))
ax = fig.add_axes([0,0,1,1])
for i in range(1,10):
    ax.plot(x,x**i, label = f'X raised to power {i}')
ax.legend(loc = 0)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='green',linewidth = 1, linestyle = '-', marker = 'o',markersize=10, 
        markerfacecolor='red')
ax.set_xlim([0,2])
ax.set_ylim([0,2])
```
